```python
# imports
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
from math import ceil

# --- configuración de paths ---
output_dir = "/Users/user/Desktop/breast_cancer_fr/plots"
os.makedirs(output_dir, exist_ok=True)

# --- carga y limpieza inicial ---
full_data = pd.read_csv("/Users/user/Desktop/breast_cancer_fr/data/processed_data.csv")

# crear columna binaria: contiene trastuzumab (1 = sí, 0 = no)
full_data["trastuzumab"] = full_data["treatment"].str.contains("trastuzumab", case=False, na=False).astype(int)

# eliminar columnas no deseadas
for col in ["index", "treatment"]:
    if col in full_data.columns:
        full_data.drop(columns=[col], inplace=True)

# limpiar menopausal.status
if "menopausal.status" in full_data.columns:
    full_data["menopausal.status"] = full_data["menopausal.status"].str.strip().str.lower()

# --- seleccionar primeras 30 columnas después de limpieza ---
primeras_30 = full_data.copy()

# si trastuzumab fue recién creado, asegurarnos que está incluido en las 30 primeras
if "trastuzumab" not in primeras_30.columns[:30]:
    columnas_finales = list(primeras_30.columns[:29]) + ["trastuzumab"]
else:
    columnas_finales = list(primeras_30.columns[:30])

exploracion_df = primeras_30[columnas_finales].copy()

# limpiar nombres de columnas
exploracion_df.columns = (
    exploracion_df.columns
    .str.strip()
    .str.replace(" mm", "", regex=False)
    .str.replace(" ", "_")
)

# --- clasificar variables ---
cat_cols = []
num_cols = []

for col in exploracion_df.columns:
    unique_vals = exploracion_df[col].nunique()
    if pd.api.types.is_numeric_dtype(exploracion_df[col]):
        if unique_vals <= 5:
            cat_cols.append(col)  # binaria o categórica codificada
        else:
            num_cols.append(col)
    else:
        cat_cols.append(col)  # tipo objeto

# --- generar boxplots para numéricas continuas ---
plots_per_fig = 6
n_figs = ceil(len(num_cols) / plots_per_fig)

for i in range(n_figs):
    fig, axs = plt.subplots(nrows=3, ncols=2, figsize=(14, 12))
    axs = axs.flatten()
    subset = num_cols[i * plots_per_fig:(i + 1) * plots_per_fig]
    
    for j, col in enumerate(subset):
        sns.boxplot(data=exploracion_df, y=col, ax=axs[j], color="skyblue")
        axs[j].set_title(col)
    
    for k in range(j + 1, len(axs)):
        fig.delaxes(axs[k])

    fig.tight_layout()
    fig.savefig(os.path.join(output_dir, f"boxplots_numericas_grupo_{i+1}.png"), dpi=300)
    plt.close(fig)

# --- generar countplots para categóricas o binarias ---
plots_per_fig = 4
n_figs = ceil(len(cat_cols) / plots_per_fig)

for i in range(n_figs):
    fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(14, 10))
    axs = axs.flatten()
    subset = cat_cols[i * plots_per_fig:(i + 1) * plots_per_fig]

    for j, col in enumerate(subset):
        sns.countplot(data=exploracion_df, x=col, ax=axs[j], palette="Set2")
        axs[j].set_title(col)
        axs[j].tick_params(axis='x', rotation=45)

    for k in range(j + 1, len(axs)):
        fig.delaxes(axs[k])

    fig.tight_layout()
    fig.savefig(os.path.join(output_dir, f"countplots_categoricas_grupo_{i+1}.png"), dpi=300)
    plt.close(fig)
```

```python
# --- matriz de correlación (Pearson) ---
# unimos numéricas continuas + binarias para correlación
corr_vars = num_cols + cat_cols  # asumen cat_cols incluye binarias como 0/1
corr_df = exploracion_df[corr_vars].copy()

# convertir todas a numéricas por si hay categóricas como 'menopausal.status'
for col in corr_df.columns:
    if not pd.api.types.is_numeric_dtype(corr_df[col]):
        corr_df[col] = corr_df[col].astype("category").cat.codes

# calcular matriz de correlación
corr_matrix = corr_df.corr(method="pearson").round(2)

# --- heatmap ---
plt.figure(figsize=(18, 14))
sns.heatmap(corr_matrix, annot=True, cmap="vlag", center=0, fmt=".2f", linewidths=0.5)
plt.title("Matriz de correlación (primeras 30 columnas limpias)", fontsize=16)
plt.tight_layout()

# guardar imagen
correlation_path = os.path.join(output_dir, "matriz_correlacion_primeras_30.png")
plt.savefig(correlation_path, dpi=300)
plt.close()
```

# Selección de features genéticas. 
Dado que tenemos más de 50k columnas, debemos seleccionar las columnas más relevantes.

```python
from sklearn.feature_selection import VarianceThreshold
import pandas as pd

# cargar dataset completo
full_data = pd.read_csv("/Users/user/Desktop/breast_cancer_fr/data/processed_data.csv")

# extraer solo las columnas de expresión génica (a partir de la 31)
gene_data = full_data.iloc[:, 30:].copy()

# eliminar genes con varianza menor a 0.01
selector = VarianceThreshold(threshold=0.01)
genes_filtrados = selector.fit_transform(gene_data)

# reconstruir dataframe con nombres de genes que quedaron
genes_filtrados_df = pd.DataFrame(genes_filtrados, columns=gene_data.columns[selector.get_support()])

# guardar si quieres
genes_filtrados_df.to_csv("/Users/user/Desktop/breast_cancer_fr/data/genes_filtrados_varianza.csv", index=False)

print(f"✅ genes seleccionados por varianza: {genes_filtrados_df.shape[1]} de {gene_data.shape[1]}")
```

```python
from sklearn.feature_selection import SelectKBest, f_classif

X = genes_filtrados_df
y = full_data["pcr_binary"]

# seleccionar los 300 genes más relevantes
selector = SelectKBest(score_func=f_classif, k=300)
X_selected = selector.fit_transform(X, y)

# nombres de los genes seleccionados
selected_genes = genes_filtrados_df.columns[selector.get_support()]
X_selected_df = pd.DataFrame(X_selected, columns=selected_genes)
X_selected_df
```

```python
# 1. seleccionar columnas clínicas limpias
clinicas_df = full_data.iloc[:, :30].copy()

# eliminar columnas no deseadas
for col in ["index", "treatment"]:
    if col in clinicas_df.columns:
        clinicas_df.drop(columns=[col], inplace=True)

# asegurarse que pcr_binary esté al final
if "pcr_binary" not in clinicas_df.columns:
    clinicas_df["pcr_binary"] = full_data["pcr_binary"]

# 2. concatenar clínico + genético (solo si tienen el mismo orden de muestras)
df_final = pd.concat([clinicas_df.reset_index(drop=True), X_selected_df.reset_index(drop=True)], axis=1)

# 3. tirar los id y guardar
df_final = df_final.drop(columns=["centerid", "patid"])
df_final.to_csv("/Users/user/Desktop/breast_cancer_fr/data/final_dataset_clinical_genetic.csv", index=False)

print(f"✅ dataset final guardado. Tamaño: {df_final.shape}")
```

### Explicación de la selección de genes

El conjunto de datos original contenía más de 54.000 genes, lo que genera múltiples inconvenientes habituales en estudios transcriptómicos:

- Hay muchas más variables que muestras, lo que puede afectar negativamente al rendimiento de los modelos.
- Existe un alto riesgo de sobreajuste si se utilizan todos los genes sin filtrado.
- El coste computacional de trabajar con todas las variables es elevado y poco eficiente.

Para reducir la dimensionalidad se aplicó un pipeline estándar de selección de características en datos ómicos, compuesto por los siguientes pasos:

**1. Filtrado por varianza**  
Se eliminaron los genes cuya expresión presentaba muy poca variación entre pacientes, al considerarse poco informativos. Para ello se aplicó un umbral de varianza mínima (0.01) mediante `VarianceThreshold`, resultando en la eliminación de un único gen.

**2. Selección supervisada con `SelectKBest`**  
Se utilizó el test ANOVA F (`f_classif`) para evaluar estadísticamente la asociación entre cada gen y la variable respuesta (`pcr_binary`). Se seleccionaron los 300 genes más relevantes, generando una nueva matriz de características con 154 muestras y 300 variables (`X_selected_df`).

---

### Explicación de la selección de variables clínicas

Las primeras 30 columnas del dataset correspondían a variables clínicas y morfológicas. No todas resultaban útiles para el análisis, por lo que se aplicó una depuración en varias fases:

**1. Eliminación de columnas irrelevantes**  
Se descartaron columnas que no aportaban valor predictivo, como:
- `index`, por tratarse de un identificador.
- `treatment`, reemplazada por una versión codificada binaria (`trastuzumab`).

**2. Normalización de categorías**  
Se estandarizó el formato de variables categóricas, como `menopausal.status`, convirtiendo los valores a minúsculas. Las variables binarias (por ejemplo, `er`, `pr`, `her2`, `ecog`, `inflammatory.brca`) se mantuvieron como enteros (0 o 1).

**3. Revisión del valor predictivo**  
Se revisaron las correlaciones entre variables clínicas usando una matriz de correlación. En función de esta revisión, se identificaron variables muy redundantes que podrían eliminarse en fases posteriores. Se conservaron aquellas con posible valor predictivo para su uso junto con los datos génicos.

---

El objetivo final es construir un dataset limpio y compacto que combine las variables clínicas relevantes y los genes seleccionados, para entrenar modelos predictivos que anticipen la respuesta patológica completa (`pcr_binary`).
