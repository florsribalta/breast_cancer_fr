```python
#imports
import pandas as pd  # type: ignore
from sklearn.feature_selection import SelectKBest, f_classif # type: ignore
from sklearn.feature_selection import VarianceThreshold, SelectKBest, f_classif  # type: ignore
from statsmodels.stats.outliers_influence import variance_inflation_factor # type: ignore
import numpy as np # type: ignore
import matplotlib.pyplot as plt # type: ignore
import os
import seaborn as sns
import numpy as np
```

```python
full_data = pd.read_csv("/Users/user/Desktop/breast_cancer_fr/data/final_dataset_clinical_genetic.csv")
full_data = full_data.drop(columns=["pcr", "pcr_binary.1"], errors="ignore")
full_data.head()
```

```python
# --- 1. revisar valores nulos ---
nulls = full_data.isnull().sum()
nulls = nulls[nulls > 0]

if nulls.empty:
    print("✅ no hay valores nulos.")
else:
    print("⚠️ columnas con nulos:\n", nulls)

# --- 2. verificar tipo de variable objetivo ---
print("\n🎯 distribución de la variable objetivo (pcr_binary):")
print(full_data["pcr_binary"].value_counts(normalize=True))

# --- 3. verificar tipos de datos ---
print("\n📦 tipos de datos:")
print(full_data.dtypes.value_counts())

# --- 4. resumen estadístico de variables numéricas ---
print("\n📊 resumen de variables numéricas:")
display(full_data.describe().T)

# --- 5. heatmap de correlaciones clínicas (y guardado) ---
import matplotlib.pyplot as plt
import seaborn as sns
import os

# seleccionar variables clínicas numéricas
clinicas_num = full_data.iloc[:, :26].select_dtypes(include=["float64", "int64"])

# matriz de correlación
corr = clinicas_num.corr()

# crear carpeta si no existe
output_dir = "/Users/user/Desktop/breast_cancer_fr/plots"
os.makedirs(output_dir, exist_ok=True)

# graficar y guardar
plt.figure(figsize=(16, 12))
sns.heatmap(corr, cmap="coolwarm", center=0, annot=False)
plt.title("Matriz de correlación - Variables clínicas")
plt.tight_layout()
plt.savefig(os.path.join(output_dir, "matriz_correlacion_clinicas.png"), dpi=300)
plt.show()
```

### Análisis exploratorio previo al modelado

Antes de entrenar los modelos predictivos, se realizaron varias comprobaciones sobre el dataset final con el objetivo de validar su calidad y adecuación para el análisis:

#### 1. Revisión de valores nulos  
Se inspeccionaron todas las columnas del dataframe para identificar la presencia de valores faltantes que pudieran requerir tratamiento, ya sea mediante imputación o eliminación.

#### 2. Distribución de la variable objetivo (`pcr_binary`)  
Se analizó la proporción de clases (0 y 1) para evaluar si existía desbalance. Esta información es relevante para decidir si es necesario aplicar técnicas como el ajuste de pesos en los modelos o el re-muestreo de las clases.

#### 3. Tipos de datos  
Se comprobaron los tipos de datos de las variables para asegurar que estuvieran correctamente codificadas como `int`, `float` o `category`. Esto es especialmente importante en el caso de variables binarias o categóricas.

#### 4. Estadísticas descriptivas  
Se revisaron estadísticas como media, desviación estándar, valores mínimos y máximos para todas las variables numéricas. El objetivo fue detectar posibles valores atípicos o diferencias de escala que pudieran afectar al modelado.

#### 5. Matriz de correlación de variables clínicas  
Se generó un mapa de calor (heatmap) con la matriz de correlación entre las variables clínicas. Esta visualización permitió:

- Identificar posibles relaciones lineales fuertes entre predictores,
- Detectar casos de multicolinealidad,
- Sugerir la reducción de variables redundantes,
- Comprender qué variables clínicas podrían estar más asociadas con la respuesta `pcr_binary`.

# Codificamos algunas columnas para poder hacer los modelos:

```python
# --- identificar columnas categóricas tipo string ---
categorical_cols = full_data.select_dtypes(include=["object"]).columns.tolist()
print(f"📦 columnas categóricas detectadas: {categorical_cols}")

# --- codificar menopausal.status con categorías ordinales ---
if "menopausal.status" in full_data.columns:
    full_data["menopausal.status"] = full_data["menopausal.status"].str.strip().str.lower()
    full_data["menopausal_status_encoded"] = full_data["menopausal.status"].astype("category").cat.codes
    full_data.drop(columns=["menopausal.status"], inplace=True)

# --- codificar race con one-hot encoding ---
if "race" in full_data.columns:
    full_data = pd.get_dummies(full_data, columns=["race"], drop_first=True)

# --- codificar arm con one-hot encoding ---
if "arm" in full_data.columns:
    full_data = pd.get_dummies(full_data, columns=["arm"], drop_first=True)

# verificación final
print(f"✅ dataset listo. Columnas actuales: {full_data.shape[1]}")
```
