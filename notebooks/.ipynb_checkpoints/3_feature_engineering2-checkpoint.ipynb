```python
full_data = pd.read_csv("/Users/user/Desktop/breast_cancer_fr/data/final_dataset_clinical_genetic.csv")
full_data = full_data.drop(columns=["pcr", "pcr_binary.1"], errors="ignore")
full_data.head()
```

```python
# --- 1. revisar valores nulos ---
nulls = full_data.isnull().sum()
nulls = nulls[nulls > 0]

if nulls.empty:
    print("âœ… no hay valores nulos.")
else:
    print("âš ï¸ columnas con nulos:\n", nulls)

# --- 2. verificar tipo de variable objetivo ---
print("\nðŸŽ¯ distribuciÃ³n de la variable objetivo (pcr_binary):")
print(full_data["pcr_binary"].value_counts(normalize=True))

# --- 3. verificar tipos de datos ---
print("\nðŸ“¦ tipos de datos:")
print(full_data.dtypes.value_counts())

# --- 4. resumen estadÃ­stico de variables numÃ©ricas ---
print("\nðŸ“Š resumen de variables numÃ©ricas:")
display(full_data.describe().T)

# --- 5. heatmap de correlaciones clÃ­nicas (y guardado) ---
import matplotlib.pyplot as plt
import seaborn as sns
import os

# seleccionar variables clÃ­nicas numÃ©ricas
clinicas_num = full_data.iloc[:, :26].select_dtypes(include=["float64", "int64"])

# matriz de correlaciÃ³n
corr = clinicas_num.corr()

# crear carpeta si no existe
output_dir = "/Users/user/Desktop/breast_cancer_fr/plots"
os.makedirs(output_dir, exist_ok=True)

# graficar y guardar
plt.figure(figsize=(16, 12))
sns.heatmap(corr, cmap="coolwarm", center=0, annot=False)
plt.title("Matriz de correlaciÃ³n - Variables clÃ­nicas")
plt.tight_layout()
plt.savefig(os.path.join(output_dir, "matriz_correlacion_clinicas.png"), dpi=300)
plt.show()
```

### AnÃ¡lisis exploratorio previo al modelado

Antes de entrenar los modelos predictivos, se realizaron varias comprobaciones sobre el dataset final con el objetivo de validar su calidad y adecuaciÃ³n para el anÃ¡lisis:

#### 1. RevisiÃ³n de valores nulos  
Se inspeccionaron todas las columnas del dataframe para identificar la presencia de valores faltantes que pudieran requerir tratamiento, ya sea mediante imputaciÃ³n o eliminaciÃ³n.

#### 2. DistribuciÃ³n de la variable objetivo (`pcr_binary`)  
Se analizÃ³ la proporciÃ³n de clases (0 y 1) para evaluar si existÃ­a desbalance. Esta informaciÃ³n es relevante para decidir si es necesario aplicar tÃ©cnicas como el ajuste de pesos en los modelos o el re-muestreo de las clases.

#### 3. Tipos de datos  
Se comprobaron los tipos de datos de las variables para asegurar que estuvieran correctamente codificadas como `int`, `float` o `category`. Esto es especialmente importante en el caso de variables binarias o categÃ³ricas.

#### 4. EstadÃ­sticas descriptivas  
Se revisaron estadÃ­sticas como media, desviaciÃ³n estÃ¡ndar, valores mÃ­nimos y mÃ¡ximos para todas las variables numÃ©ricas. El objetivo fue detectar posibles valores atÃ­picos o diferencias de escala que pudieran afectar al modelado.

#### 5. Matriz de correlaciÃ³n de variables clÃ­nicas  
Se generÃ³ un mapa de calor (heatmap) con la matriz de correlaciÃ³n entre las variables clÃ­nicas. Esta visualizaciÃ³n permitiÃ³:

- Identificar posibles relaciones lineales fuertes entre predictores,

# Codificamos algunas columnas para poder hacer los modelos:

```python
# --- identificar columnas categÃ³ricas tipo string ---
categorical_cols = full_data.select_dtypes(include=["object"]).columns.tolist()
print(f"ðŸ“¦ columnas categÃ³ricas detectadas: {categorical_cols}")

# --- codificar menopausal.status con categorÃ­as ordinales ---
if "menopausal.status" in full_data.columns:
    full_data["menopausal.status"] = full_data["menopausal.status"].str.strip().str.lower()
    full_data["menopausal_status_encoded"] = full_data["menopausal.status"].astype("category").cat.codes
    full_data.drop(columns=["menopausal.status"], inplace=True)

# --- codificar race con one-hot encoding ---
if "race" in full_data.columns:
    full_data = pd.get_dummies(full_data, columns=["race"], drop_first=True)

# --- codificar arm con one-hot encoding ---
if "arm" in full_data.columns:
    full_data = pd.get_dummies(full_data, columns=["arm"], drop_first=True)

# verificaciÃ³n final
print(f"âœ… dataset listo. Columnas actuales: {full_data.shape[1]}")
```
